```python
from pathlib import Path
from cobra.io import load_json_model,load_model
from cobra.io import write_sbml_model
from cameo.flux_analysis import pfba
from cameo.strain_design import FSEOF
```

```python
# get load path
loadPath = Path(" ") / ".."  / "" / "F1515.json"
loadPath = loadPath.resolve()
model=load_json_model(str(loadPath))
model
```

```python
write_sbml_model(model,'F1515.xml')
```

```python
# compute pFBA algorithm
pFBAsolution=pfba(model=model,objective=model.reactions.EX_15dap_e)
```

```python
pFBAsolution.fluxes['EX_15dap_e']
```

```python
pFBAsolution.data_frame.to_csv("result.csv")
```

```python
fseof = FSEOF(
    model=model,
    # enforced_reaction=model.reactions.EX_15ptnd_e,
    primary_objective=model.reactions.BIOMASS_Ec_iML1515_core_75p37M,
)
fseof_solution=fseof.run(
    target=model.reactions.EX_15dap_e,
    max_enforced_flux=0.9,
    number_of_results=100
)
```
